# A simple example

Let's first create a simple dataset and make a figure! Here are the basic components of a matplotlib visualization:

**Figure:** The entire plot or canvas for the visualization.

**Axes:** The area/pane within a figure where you plot the data. You can have multiple axes in one figure.

**Line & Marks:** Data representations

**Axis:** X & Y, with limits and ticks representing data

**Labels:** Labels for X & Y axes, labels for data points

**Title:** Text at the top of the plot

```python
# Create a blank canvas

# Create a plot
fig, ax = plt.subplots()

# Show the plot
plt.show()
```

```python
# Create a plot
fig, ax = plt.subplots()

# Create some data
x = [4.0, 5.0, 6.0, 7.0, 8.0, 9.0, 10.0, 11.0, 12.0, 13.0, 14.0]
y = [4.26, 6.95, 7.24, 4.82, 9.96, 7.58, 8.81, 8.04, 8.33, 10.84, 5.68,]

# Plot the data on the axis as a scatter plot
ax.scatter(x, y)

# Optionally save the figure as a PDF file or PNG image
plt.savefig('test.pdf')
plt.savefig('test.png')

# Show the plot
plt.show()
```

# Mapping data to mark properties

### Plotting function take parameters that given the style of those marks

```python

fig, ax = plt.subplots()

# First, change the default size and color of the scatter plot
ax.scatter(x, y, s=100, color='red', marker='*')

# For more colors, see: https://matplotlib.org/stable/gallery/color/named_colors.html
# you can also use hex codes like color="#058ED9" or RGB values color=[0.8, 0.5, 0.1]

# For marker styles, see: https://matplotlib.org/stable/api/markers_api.html

plt.show()
```

## Now map data to these mark properties

```python

# Add a third dimension of data
z = np.array([10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0])

fig, axes = plt.subplots(nrows=2, figsize=(4, 8))

# make a scatter plot, use the z variable to control size
ax = axes[0]
ax.scatter(x, y, s=z*8)

# or use the z variable to control color
ax = axes[1]
ax.scatter(x, y, c=z)

plt.show()
```

## Different functions take some different arguments; check the documentation for details

```python

fig, ax = plt.subplots()

# Change the style of a line plot
ax.plot(x, y, linewidth=4, linestyle='--', alpha=0.5)

# See function definition for full details:
### https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

plt.show()
```

## Modify the non-data elements of the plot (e.g., axes, etc.)

```python
fig, ax = plt.subplots()

ax.scatter(x, y)

# Change the limits of the y-axis
ax.set_ylim(0, 13)

# Change the tick-marks on x-axis
ax.set_xticks([4, 9, 14])

# Add a label on the x-axis
ax.set_xlabel("x-axis")

# Add a title:
ax.set_title("Title")


# Show the plot
plt.show()
```

## More complex plots, strating by example

### Take a look at the matplotlib example gallery: https://matplotlib.org/stable/gallery/index.html


```python
# Let's try making a contour plot: https://matplotlib.org/stable/gallery/images_contours_and_fields/contour_label_demo.html

# Generate some data
delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

# Make the canvas
fig, ax = plt.subplots()

# Basic contour plot
CS = ax.contour(X, Y, Z)

# Add the labels
ax.clabel(CS, CS.levels, fontsize=10)

plt.show();
```

```python
# Now make the same plot as a heatmap: https://matplotlib.org/stable/gallery/images_contours_and_fields/image_annotated_heatmap.html

fig, ax = plt.subplots()

# Plot the Z values as a heatmap, flipping the order of the rows because of the imshow format
ax.imshow(Z[::-1], cmap='coolwarm')

plt.show();
```
