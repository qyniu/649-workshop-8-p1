# Simple Example

Let's recall: *Data + Mark + Encoding = Visualization*

Let's start with a simple scatter plot using data from Altair sample dataset

```python
#Import datasets from vega_datasets
from vega_datasets import data

#Retrieve cars dataset and review it
cars = data.cars()
cars.head()
```

```python
# Let's specify mark type for the chart
chart1 = alt.Chart(cars).mark_point()
chart1
```

Without encoding channels defined, Altair collapses all observations in the dataset into one aggregated point displayed on the chart.

That said, if we discuss the code part by part, we're essentially:

1. Creating an Altair chart object from our `cars` dataset using `alt.Chart(cars)`
2. We're then chaining the `mark_point()` method to specify that our chart should use point marks (scatter plot)
3. Finally, we're assigning our chart to a variable `chart` and displaying it

```python
# TODO: Try encoding the dataset using circles
# If you want to assign your chart to a variable, use 'chart2'
# Map columns to x, y, and color
chart1 = alt.Chart(cars).mark_point().encode(
    x='Horsepower:Q',    # :Q tells Altair this is Quantitative (numeric)
    y='Miles_per_Gallon:Q',
    color='Origin:N',     # :N tells Altair this is Nominal (categorical)
    tooltip=['Name', 'Year']
).interactive()

chart1.display()
```

^ Hint: mark_circle()

```python
# Now let's add one encoding
chart1 = alt.Chart(cars).mark_point().encode(
    x = 'Horsepower'
)
chart1
```

We're now adding an encoding channel by chaining the `.encode(x='Horsepower')` method. This maps the `Horsepower` column to the x-axis, so each data point is positioned horizontally based on its horsepower value.

```python
# TODO: Try to map 'Acceleration' to x-axis
# If you want to assign your chart to a variable, use 'chart2'
chart1 = alt.Chart(cars).mark_point().encode(
    x='Horsepower:Q',
    y='Miles_per_Gallon:Q',
    color='Origin:N',  # Colors the points by where the car was made
    shape='Origin:N'  # Changes the point shape for better accessibility
).interactive()

chart1.display()
```

```python
# Finally, let's add one more encoding channel
chart1 = alt.Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon'
)

chart1
```

Observations from the example:


*   Altair automatically inferred data types from the pandas DataFrame
*   Created appropriate scales
*   Generated axis labels

# Data Type Specification



When using data other than Pandas DataFrames, it is important to manually specify the data types, as it affects the visualization. Let's see the example:

```python
#Base chart
base = alt.Chart(cars).mark_point().encode(
    x='Horsepower:Q',
    y='Miles_per_Gallon:Q'
)

#Add one more channel to the base chart to see how different data types impact the visualization
quantitative = base.encode(color='Cylinders:Q').properties(title='Quantitative')
ordinal = base.encode(color='Cylinders:O').properties(title='Ordinal')
nominal = base.encode(color='Cylinders:N').properties(title='Nominal')

#Show side by side
quantitative | ordinal | nominal
```

Recall from slides that:


*   Q - quantitative (continuous real-valued quantity)
*   O - ordinal (discrete ordered quantity)
*   N - nominal (discrete unordered category)
*   T - temporal (time or date)
*   G - geojson (geographic shape)


# Encoding Channels


Let's explore some of the most common encoding channels that Altair supports

```python
alt.Chart(cars).mark_point().encode(
    x='Horsepower:Q',
    y='Miles_per_Gallon:Q',
    color='Origin:N',
    # size='Cylinders:Q',
    tooltip=['Name', 'Horsepower', 'Miles_per_Gallon', 'Origin']
).properties(
    title='Car Characteristics by Origin'
)
```

Recall from slides that some of the common encoding channels include:

* Position: x, y
* Mark properties: color, shape, size, opacity, etc
* Text: text, tooltip


# Mark Types

Altair offers a range of different mark types for different purposes. Let's review some of them:

```python
# Scatter plot showing the relationship between horsepower and MPG, colored by origin
alt.Chart(cars).mark_circle().encode(
    x='Horsepower:Q',
    y='Miles_per_Gallon:Q',
    color='Origin:N',
    tooltip=['Name', 'Horsepower', 'Miles_per_Gallon', 'Origin']
).properties(
    title='Car Characteristics by Origin'
)
```

```python
# Tick plot showing horsepower distribution across different car origins
alt.Chart(cars).mark_tick().encode(
    x='Horsepower:Q',
    y='Origin:N',
    color='Origin:N'
).properties(title='Horsepower Distribution by Origin')
```

```python
# Line chart showing the trend in average miles per gallon over time
alt.Chart(cars).mark_line().encode(
    x='Year:T',
    y='mean(Miles_per_Gallon):Q'
).properties(
    title='Average MPG Trend Over Time'
)
```

```python
#Bar chart showing the number of cars from different countries
alt.Chart(cars).mark_bar().encode(
    x='count():Q',
    y='Origin:N',
    tooltip=['Origin:N', 'count():Q']
).properties(
    title='Number of Cars by Origin'
)
```

```python
# Pie chart showing the distribution of cars by number of cylinders
alt.Chart(cars).mark_arc().encode(
    theta='count():Q',
    color='Cylinders:N',
    tooltip=['Cylinders:O', 'count():Q']
).properties(
    title='Cars by Number of Cylinders'
)
```

```python
# Heat map showing the distribution of cars by weight and acceleration
alt.Chart(cars).mark_rect().encode(
    x=alt.X('Acceleration:Q').bin(maxbins=10),
    y=alt.Y('Weight_in_lbs:Q').bin(maxbins=10),
    color='count():Q'
).properties(
    title='Distribution of Cars by Acceleration and Weight'
)
```
